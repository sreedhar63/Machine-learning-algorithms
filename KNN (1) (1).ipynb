```python
from sklearn.neighbors import KNeighborsClassifier

# Features: all columns except 'animal_name' and perhaps 'type' (if that's the label)
X = zoo.drop(['animal name', 'type'], axis=1)
y = zoo['type']  # or any column you want as the target

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

# Filter to top features
selector = SelectKBest(mutual_info_classif, k=5)

# Build pipeline
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('select', selector),
    ('knn', KNeighborsClassifier())
])

# Cross-validate for different k values
scores = []
for k in k_range:  # Iterate through k_range instead of score
    pipe.set_params(knn__n_neighbors=k)
    cv = cross_val_score(pipe, X, y, cv=5, scoring='accuracy')
    scores.append(np.mean(cv))

best_k = k_range[np.argmax(scores)] # Get the corresponding k from k_range
print("Best k:", best_k, "with CV accuracy:", max(scores))
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
from itertools import combinations
from sklearn.neighbors import KNeighborsClassifier

h = 0.1
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ['#FF0000', '#00FF00', '#0000FF']

# Refit the pipeline with the best_k found
pipe.set_params(knn__n_neighbors=best_k)
pipe.fit(X, y)

# Transform X to get the selected features for plotting
X2 = pipe.named_steps['select'].transform(X)
y2 = y # y remains the same

clf = KNeighborsClassifier(n_neighbors=best_k)
clf.fit(X2, y2)

# Meshgrid - use X2 for min/max and correct indexing
x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1 # Use X2[:, 1]
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))



plt.figure()

# Scatter plot - use X2 for plotting the data points
plt.scatter(X2[:, 0], X2[:, 1], c=y2, cmap=ListedColormap(cmap_bold), edgecolor='k') # Use X2 and y2
plt.title(f"KNN decision boundary (k={best_k})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
